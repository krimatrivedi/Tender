```python
''' A binary tree node has data, pointer to left child
and a pointer to right child '''
class newNode:
	def __init__(self, data):
		self.data = data
		self.left = None
		self.right = None

''' Helper function that allocates a new node with the
given data and NULL left and right pointers. '''
def output(head):
	
	temp = head
	visited = set()
	while (temp and temp not in visited):
		
		if (temp.left and temp.left not in visited):
			temp = temp.left
			
		elif (temp.right and temp.right not in visited):
			temp = temp.right
		
		else:
			print(temp.data, end = " ")
			visited.add(temp)
			temp = head

if __name__ == '__main__':
	
	input = newNode(1)
	input.left = newNode(2)
	input.right = newNode(3)
	input.left.left = newNode(4)
	input.left.right = newNode(5)
	input.left.right.left = newNode(6)
	input.left.right.right = newNode(7)
	input.right.right = newNode(8)
	input.right.right.left = newNode(9)
	output(input)
```
